```python
import requests  # Importamos el módulo requests para realizar solicitudes 
                 # HTTP

res = requests.get('https://pokedex-api-server.onrender.com/api/v1/pokedex')
# Realizamos una solicitud GET a la API

if res:  # Verificamos si la solicitud fue exitosa
    print("OK")  # Imprimimos "OK" si la solicitud fue exitosa
else:
    print("Response Failed")  # Imprimimos "Response Failed" si la 
                              # solicitud falló
```

```python
import requests

url = 'https://pokedex-api-server.onrender.com/api/v1/pokedex?limit=100'

#url_original https://pokedex-api-server.onrender.com/api/v1/pokedex
#Se puede cambiar la paginacion de los pokemons a mostrar y la pagina 
#posicionada a mostrar

params = dict(texT="Hello", lang="en-es")

res = requests.get(url, params=params)

json = res.json()

#Para acceder la "ruta" es json["docs"][1] donde 1 es el pokemon con sus
#respectivos datos

print("100 Primeros Pokemons por ID")
for pokemon in json["docs"]:
    print("Pokemon id:",pokemon["_id"],", ",pokemon["nombre"])
```

```python
from PIL import Image  # Importamos la clase Image de la biblioteca PIL
from io import BytesIO  # Importamos el módulo BytesIO para leer datos 
                        # binarios en memoria
from IPython.display import display #Importamos la clase de la biblioteca
                                    # IPython
import requests  # Importamos el módulo requests para realizar solicitudes
                 # HTTP


url = 'https://assets.pokemon.com/assets/cms2/img/pokedex/full/491.png'  
# URL de la imagen del Pokémon Darkrai

response = requests.get(url)  # Realizamos una solicitud GET a la URL 
                              # para obtener los datos de la imagen
image = Image.open(BytesIO(response.content)) 
# Abrimos y cargamos la imagen desde los datos binarios obtenidos

display(image) # Mostramos la imagen en una ventana emergente

#Posdata: Mi pokemon favorito
```
